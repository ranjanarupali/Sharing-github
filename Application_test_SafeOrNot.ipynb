```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import json
%matplotlib inline 
from sklearn.model_selection import train_test_split

#opening JSON file
f = open('Downloads/assignment.json')

data = json.load(f)
df = pd.DataFrame(data)

df.head(100)
```

```python
#cleaning the dataframe
bn = pd.DataFrame(df.req.values.tolist())
bn.head()
```

```python
# dropping the column which are not used because the each column has the same values.

total_df= total_df.drop(['baseUrl','fresh','hostname','ip','ips','originalUrl','params','path','protocol','query','secure','stale','subdomains','xhr','host','user-agent','content-type','org_id','user_session_id','accept','content-length','name','role'],axis=1)
```

```python
total_df=pd.get_dummies(total_df,columns=['isSafe','title','desc'],drop_first=True)

total_df
```

```python
##Assigning the values of x and y

X = total_df.drop(['isSafe_True'],axis=1).values
y = total_df['isSafe_True'].values
```

```python
#Next, we'll split the data into training and test datasets

from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=.3, random_state=0)

print('There are {} samples in the training set and {} samples in the test set'.format(
X_train.shape[0], X_test.shape[0]))
print()
```

```python
from sklearn.svm import SVC

clf_SVC = SVC(C=100.0, kernel='rbf', degree=3, gamma='auto', coef0=0.0, shrinking=True, 
          probability=False, tol=0.001, cache_size=200, class_weight=None, 
          verbose=0, max_iter=-1, decision_function_shape="ovr", random_state = 0)
clf_SVC.fit(X_train,y_train)

print('Accuracy of SVC on training set: {:.2f}'.format(clf_SVC.score(X_train, y_train) * 100))

print('Accuracy of SVC on test set: {:.2f}'.format(clf_SVC.score(X_test, y_test) * 100))
```

```python
# Fitting Decision Tree Classifier to the Training set
from sklearn.tree import DecisionTreeClassifier
classifier = DecisionTreeClassifier()
print(classifier)
classifier.fit(X_train, y_train)

# Predicting the Test set results
y_pred = classifier.predict(X_test)

# Accuracy on the Test set results                              
from sklearn.metrics import accuracy_score
print('\n'+'-'*20+'Accuracy Score on the Test set'+'-'*20)                             
print("{:.0%}".format(accuracy_score(y_test,y_pred)))
```

```python
# Fitting Random Forest Classifier to the Training set
from sklearn.ensemble import RandomForestClassifier
classifier = RandomForestClassifier()
print(classifier)
classifier.fit(X_train, y_train)

# Predicting the Test set results
y_pred = classifier.predict(X_test)

# Accuracy on the Test set results                              
from sklearn.metrics import accuracy_score
print('\n'+'-'*20+'Accuracy Score on the Test set'+'-'*20)                             
print("{:.0%}".format(accuracy_score(y_test,y_pred)))
```
